## Decision Trees

Decision trees are widely used models **for classification and regression** tasks. Essentially, they learn a hierarchy of if/else questions, leading to a decision.

```python
import sklearn
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
```

```python
iris = load_iris()
X = iris.data[:,2:]
y = iris.target
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

```python
image_path = "C:\\Users\\prajeeth.nair\\Desktop\\iris_tree.dot"
```

```python
export_graphviz(
        tree_clf,
        out_file="image_path",
        feature_names=iris.feature_names[2:],
        class_names=iris.target_names,
        rounded=True,
        filled=True
    )
```

One of the many qualities of Decision Trees is that they require very little data preparation. They dont really require feature scaling or centering at all.
